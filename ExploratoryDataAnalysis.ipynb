```python
# import pandas and numpy library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import datetime as dt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# import or load the json file and convert it to pandas dataframe
logins = pd.read_json('logins.json')
```

```python
logins.head()
```

```python
# check the infromation from the dataframe
logins.info()
```

```python
# lets sort the dataframe
logins = logins.sort_values('login_time')
```

```python
logins.head()
```

```python
# join the dataframe with a new dataframe which include year month day hour minutes and second column
logins = logins.join(pd.DataFrame({'y': logins.login_time.dt.year, 'm':logins.login_time.dt.month, 'd':logins.login_time.dt.day, 'h':logins.login_time.dt.hour, 'minu':logins.login_time.dt.minute, 'se':logins.login_time.dt.second}))
```

```python
logins.head()
```

```python
logins.tail()
```

```python
#create a column named interval_time where we store if a login happened in 1st, 2nd, 3rd or 4th quarter of 15 minute time inteval
logins['interval_time'] = np.where(logins['minu']<15, 0, np.where(logins['minu']<30,15,np.where(logins['minu']<45,30,np.where(logins['minu']<60,45,logins.minu))))
```

```python
logins.head()
```

```python
logins.info()
```

```python
# create a new column named data where we store datatime object with year/month/data
logins['date'] = pd.to_datetime(dict(year=logins.y,month=logins.m,day=logins.d))
```

```python
logins.head()
```

```python
# add hour as a time to the date column and create a new column
logins['date_interval_h'] = logins['date']+pd.to_timedelta(logins['h'], unit='h')
```

```python
# add minute as time to the date_interval_h column and create a new column
logins['date_interval'] = logins['date_interval_h']+pd.to_timedelta(logins['interval_time'], unit='m')
```

```python
logins.head()
```

# Grouping the data by datetime every 15 minute inteval

```python
logins_interval_time = logins.groupby(['date_interval'])['login_time'].count()
df_time = pd.DataFrame(logins_interval_time)
```

```python
df_time = pd.DataFrame(logins_interval_time)
```

```python
df_time.head()
```

```python
# change the name of the column into number_of_logins
df_time.columns = ['number_of_logins']
```

```python
# plot the graph
plt.figure(figsize=(40,10))
plt.xlabel('Date')
plt.ylabel('Number of logins')
plt.title('Time-series Graph for Number of Logins based on 15 minute time interval')
plt.plot(df_time)
plt.xticks(rotation=90)
plt.tight_layout()
```

The graph is noisy and difficult to interpret, so we can not say if there is no upward or downward trend.

```python
# plot the graph
plt.xlabel('Date')
plt.ylabel('Number of logins')
plt.title('Time-series Graph for Number of Logins based on 15 minute time interval on January 1st')
plt.plot(df_time.loc['1970-01-01'])
plt.xticks(rotation=90)
plt.tight_layout()
# the graph is for one day 1970-01-01
```

On 1970-01-01, the logins were saved starting 20:00:00. The least number of logins were at 9pm and 10:30 pm. The maximum number of logins was at 11pm.

```python
# Create a new dataframe to group the data by hour, day of week and month.
df_number_of_logins = df_time.reset_index()
```

```python
# check the dataframe
df_number_of_logins.head()
```

```python
# join the dataframe with a new dataframe which include year month day hour and minutes column
df_number_of_logins = df_number_of_logins.join(pd.DataFrame({'y': df_number_of_logins.date_interval.dt.year, 'm':df_number_of_logins.date_interval.dt.month, 'd':df_number_of_logins.date_interval.dt.day, 'h':df_number_of_logins.date_interval.dt.hour, 'minu':df_number_of_logins.date_interval.dt.minute}))
```

```python
# check the dataframe
df_number_of_logins.head()
```

# Login Pattern by Hour

```python
# Goup the data by hour to find out the average hourly logins
df_hourly_avg_logins = df_number_of_logins.groupby('h')['number_of_logins'].mean().reset_index()
```

```python
# check the dataframe 
df_hourly_avg_logins
```

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.xlabel('Hours')
plt.ylabel('Average Number of logins')
plt.title('Hourly Average Number of Logins')
sns.barplot(data=df_hourly_avg_logins,x='h',y='number_of_logins')
plt.tight_layout()
```

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.xlabel('Hours')
plt.ylabel('Average Number of logins')
plt.title('Hourly Average Number of Logins')
sns.lineplot(data=df_hourly_avg_logins, x='h', y='number_of_logins')
plt.tight_layout()
```

The bar plot and line plot shows us that the peak hours are at midnight to 1 am. Then the number of logins drops until 7am. At 7am to 8am we have the lowest average number of logins. Then, The average number of logins increases up until 11 am. Then, it follows a downward trend till 5pm. After 5pm, the average number of logins increases and follow an upward trend. 

Most of the people sleep close to midnight, and wakes up around 6-8am in the morning. This might have effected the drop in average number of logins. People start their day or go to work after 8 am till 11 am. Most of the people stay busy at work up to 5 pm. At 5pm they finish their work. 

# Logins Pattern by Day of Week

```python
import calendar
df_number_of_logins['day_of_week'] = df_number_of_logins['date_interval'].apply(lambda x: x.weekday()) # get the weekday index, between 0 and 6
df_number_of_logins['day_of_week'] = df_number_of_logins['day_of_week'].apply(lambda x: calendar.day_name[x])
df_number_of_logins.head()
```

```python
# Group the data by day_of_week to find out the average number of logins
df_day_of_week_avg_logins = df_number_of_logins.groupby('day_of_week')['number_of_logins'].mean().reset_index()
```

```python
# check the dataframe
df_day_of_week_avg_logins.sort_values('day_of_week')
```

```python
sorter = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
sorterIndex = dict(zip(sorter,range(len(sorter))))
sorterIndex
```

```python
df_day_of_week_avg_logins['day_id'] = df_day_of_week_avg_logins['day_of_week'].map(sorterIndex)
df_day_of_week_avg_logins.sort_values('day_id', inplace=True)
df_day_of_week_avg_logins
```

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.xlabel('Day of Week')
plt.ylabel('Average Number of logins')
plt.title('Day of Week Average Number of Logins')
sns.barplot(data=df_day_of_week_avg_logins,x='day_of_week',y='number_of_logins')
plt.tight_layout()
```

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.xlabel('Day of Week')
plt.ylabel('Average Number of logins')
plt.title('Day of Week Average Number of Logins')
sns.lineplot(data=df_day_of_week_avg_logins, x='day_id', y='number_of_logins')
plt.tight_layout()
```

The barplot and lineplot shows a downward trend from Sunday to Monday. Then, there is a upward trend from Monday. The maximum average number of logins is on Saturday and the minimum average number of logins is on Monday.

Most of the people finish weekend and start work week on Monday. This might have effected the drop in number of logins. Most of the people have off day on weekends, and this tells us that most of our user logins on weekends.

```python
# plot the graph
plt.figure(figsize=(10,10))
plt.xlabel('Months')
plt.ylabel('Average Number of logins')
plt.title('Monthly Average Number of Logins')
sns.lineplot(data=df_monthly_avg_logins, x='m', y='number_of_logins')
plt.tight_layout()
```

The bar plot and line plot Monthly Average Number of Logins shows us a upward trend from January till March. After March, the average number of logins has slightly decreased. We can not say there is a downward trend. We need to collect more data for following months to confirm any trend.
